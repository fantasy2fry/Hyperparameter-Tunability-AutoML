### Salary data

```python
df4 = pd.read_csv("../data/salary.csv")
df4.head()
```

```python
df4.info()
```

```python
# Must do quick cleaning
df4.apply(lambda x: (print(x.name, "\n", x.unique())) if x.dtype==np.dtype('O') else 0);
```

```python
# Binary
df4.sex = df4.sex.map(lambda x: x.strip()).map({"Male": 1, "Female": 0})
df4.salary = df4.salary.map(lambda x: x.strip()).map({"<=50K": 0, ">50K": 1})

# Oridinal
education_levels = [
    ' Preschool', ' 1st-4th', ' 5th-6th', ' 7th-8th', ' 9th', ' 10th', ' 11th', ' 12th',
    ' HS-grad', ' Some-college', ' Assoc-acdm', ' Assoc-voc', ' Bachelors', 
    ' Masters', ' Prof-school', ' Doctorate'
]

# Map in order bcs education has > relation
education_to_int = {level: idx for idx, level in enumerate(education_levels)}
df4.education = df4.education.map(education_to_int)
# Get categorical cols
non_categorical_cols = df4.select_dtypes(exclude=["object"]).columns
categorical_cols = df4.drop(non_categorical_cols, axis = 1).columns
# if value_count < ~3% of rows map to Other
for col in categorical_cols:
    val_counts = df4[col].map(lambda x: x.strip()).value_counts()
    df4[col] = df4[col].map(lambda x: x.strip()).map(lambda x: x if val_counts[str(x)] > 1000 else "Other")
# One_hot_encoding
df4 = pd.get_dummies(data=df4, columns=categorical_cols)
cols = [col for col in df4.columns if col != 'salary'] + ['salary']
df4 = df4[cols]
```

```python
df4.info()
```

```python
# Then take only 1000 observations to minimalize time spending on tuning hiperparameters...
df4 = df4.sample(1000, random_state = 17)
```

### X, y split and save

```python
Xs = []
ys = []
dfs = [df1, df2, df3, df4]
for df in dfs:
    Xs.append(df.drop(df.columns[-1], axis = 1))
    # target column is last in every dataset
    ys.append(df.iloc[:, -1])
```

```python
Xs[0].to_csv("../data/credit_X.csv", index=False)
Xs[1].to_csv("../data/raisin_X.csv", index=False)
Xs[2].to_csv("../data/alzheimer_X.csv", index=False)
Xs[3].to_csv("../data/salary_X.csv", index=False)

ys[0].to_csv("../data/credit_y.csv", index=False)
ys[1].to_csv("../data/raisin_y.csv", index=False)
ys[2].to_csv("../data/alzheimer_y.csv", index=False)
ys[3].to_csv("../data/salary_y.csv", index=False)
```
